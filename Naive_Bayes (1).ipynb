```python
# importing libraries  
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd 
from matplotlib import pyplot
import numpy as np
import matplotlib.pyplot as plt
#importing datasets  
data_set= pd.read_csv('C:/Users/asus/Desktop/my patents/pupil vs wmc project/diabetes_012_health_indicators_BRFSS2015.csv')  
  
#Extracting Independent and dependent Variable  
x= data_set.iloc[:, [4,15]].values  
y= data_set.iloc[:, 0].values
z= data_set.iloc[:, 15].values
m= data_set.iloc[:, 4].values
  
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0) 
z_train, z_test= train_test_split(z, test_size= 0.25, random_state=0) 
m_train, m_test= train_test_split(m, test_size= 0.25, random_state=0) 
  
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()  
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)
# Fitting Naive Bayes to the Training set  
from sklearn.naive_bayes import GaussianNB  
classifier = GaussianNB()  
classifier.fit(x_train, y_train) 
#Creating the Confusion matrix 
y_pred = classifier.predict(x_test)
print(y_pred)
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred) 
print ("Confusion Matrix : \n", cm)
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
from matplotlib.colors import ListedColormap
#Visulaizing the trianing set result  
from matplotlib.colors import ListedColormap  
x_set, y_set = x_train, y_train  
x1, x2 = nm.meshgrid(nm.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step  =0.01),  
nm.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))  
mtp.contourf(x1, x2, classifier.predict(nm.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),  
alpha = 0.75, cmap = ListedColormap(('purple','green' )))  
mtp.xlim(x1.min(), x1.max())  
mtp.ylim(x2.min(), x2.max())  
for i, j in enumerate(nm.unique(y_set)): 
    mtp.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],  
        c = ListedColormap(('purple', 'green'))(i), label = j)  

mtp.title('Naive Bayes Algorithm (Training set)')  
mtp.xlabel('BMI')  
mtp.ylabel('Mental Health')  
mtp.legend()  
mtp.show()
#Visualizing the predictions
from matplotlib.colors import ListedColormap
x_set1, y_set1 = x_test, y_pred  
x11, x21 = nm.meshgrid(nm.arange(start = x_set1[:, 0].min() - 1, stop = x_set1[:, 0].max() + 1, step  =0.01),  
nm.arange(start = x_set1[:, 1].min() - 1, stop = x_set1[:, 1].max() + 1, step = 0.01))  
mtp.contourf(x11, x21, classifier.predict(nm.array([x11.ravel(), x21.ravel()]).T).reshape(x11.shape),  
alpha = 0.75, cmap = ListedColormap(('purple','green' )))  
mtp.xlim(x11.min(), x11.max())  
mtp.ylim(x21.min(), x21.max())  
for i, j in enumerate(nm.unique(y_set)): 
    mtp.scatter(x_set1[y_set1 == j, 0], x_set1[y_set1 == j, 1],  
        c = ListedColormap(('purple', 'green'))(i), label = j)  

mtp.title(' Naive Bayes Algorithm (Test set)')  
mtp.xlabel('BMI')  
mtp.ylabel('Mental Health')  
mtp.legend()  
mtp.show() 
# Function to make predictions
def prediction(x_test, classifier):
  
    # Predicton on test with giniIndex
    y_pred = classifier.predict(x_test)
   
    return y_pred
# Plot of predictions with Mental health data
k=prediction(x_test, classifier)
xPoints = z_test
yPoints = k
pyplot.plot(xPoints, yPoints)
plt.ylabel('Prediction -Diabetic')
plt.xlabel('Mental Health')
pyplot.show()
# Plot of predictions with BMI
uPoints = m_test
vPoints = k
pyplot.plot(uPoints, vPoints)
plt.ylabel('Prediction -Diabetic')
plt.xlabel('BMI')
pyplot.show()
```
